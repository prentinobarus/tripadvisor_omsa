```python
import pandas as pd
import os
import glob
```

```python
path = os.getcwd() + '/Dataset_new/'
```

```python
filename = 'udaya_resort.csv'
hotelname = 'The Udaya Resort & Spa'
```

```python
df = pd.read_csv(path+filename)
df = df.drop(columns='Unnamed: 0')
df.insert(loc=0, column='hotel', value=hotelname)
df.to_csv('Dataset_new/'+filename)
df
```

```python
# Combine

hotel_csv = os.path.join(path, '*.csv')
joined_review = glob.glob(hotel_csv)

hotel_df = pd.concat(map(pd.read_csv, joined_review), ignore_index=True)
```

```python
stemed = pd.read_csv('stemmed_review.csv')
```

```python
hotel_df
```

```python
stemed
```

```python
pd.concat(g for _, g in hotel_df.groupby("review") if len(g) > 1)
```

```python
pd.concat(g for _, g in stemed.groupby("review") if len(g) > 1)
```

```python
hotel_df = hotel_df.drop_duplicates(subset=['review'])
```

```python
stemed.shape
```

```python
hotel_df = hotel_df.drop(hotel_df.columns[0], axis = 1)
stemed = stemed.drop(stemed.columns[0], axis = 1)
```

```python
hotel_df['review'] = hotel_df['review'].str.lower()
```

```python
hotel_df
```

```python
hotel_df = hotel_df.drop(hotel_df.columns[-1], axis = 1)
```

```python
hotel_df
```

```python
result = pd.merge(stemed, hotel_df, on='review')
```

```python
result
```

```python
new_cols = ['hotel', 'review', 'stars', 'tokenized', 'no_slang', 'stemmed', ]
result = result[new_cols]
```
